```python
#task1 : #Write a Python function that takes a year as input and checks whether it is a leap year.
#The function should return an appropriate message indicating whether the year is a leap year or not.
#Do not include any input-output examples.
def is_leap_year(year):
    """Checks whether a given year is a leap year."""
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return f"{year} is a leap year."
    else:
        return f"{year} is not a leap year."
# Example usage
year = 2020
print(is_leap_year(year))
year = 2021
print(is_leap_year(year))
```

```python
#task2:Generate a Python script that stores user details (name, email, password) in a file. Then analyze if storing the
#password is secure and modify the script to store the password safely using hashing instead of plain text.
import hashlib
import json
import os
from pathlib import Path
def hash_password(password):
    """Hash a password using SHA-256"""
    return hashlib.sha256(password.encode()).hexdigest()
def save_user_details(name, email, password, filename="users.json"):
    """Save user details with hashed password to a JSON file"""
    hashed_password = hash_password(password)
    
    user_data = {
        "name": name,
        "email": email,
        "password_hash": hashed_password
    }
    
    users = []
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            users = json.load(f)
    
    users.append(user_data)
    
    with open(filename, 'w') as f:
        json.dump(users, f, indent=4)
    
    print(f"User {name} saved securely!")
def verify_password(stored_hash, provided_password):
    """Verify if provided password matches the stored hash"""
    return stored_hash == hash_password(provided_password)
# Example usage
if __name__ == "__main__":
    save_user_details("John Doe", "john.doe@example.com", "securepassword123")

    save_user_details("Jane Smith", "jane.smith@example.com", "anothersecurepassword456")
    print("\nVerifying password for John Doe:")
    print(verify_password(
        "a8b64babd3f5e6c3f4e5d6c7b8a9b0c1d2e3f4a5b6c7d8e9f0a1b2c3d4e5f6a7",
        "securepassword123"
    ))
    print("\nVerifying password for Jane Smith:")
    print(verify_password(
        "b1c2d3e4f5a6b7c8d9e0f1a2b3c4d5e6f7a8b9c0d1e2f3a4b5c6d7e8f9a0b1c2",
        "wrongpassword"
    ))
```

```python
#task3
def format_name(full_name):
    """
    Format a full name as "Last, First"
    
    Args:
        full_name (str): A full name in "First Last" format
        
    Returns:
        str: The name formatted as "Last, First"
    """
    parts = full_name.split()
    if len(parts) >= 2:
        first_name = parts[0]
        last_name = parts[-1]
        return f"{last_name}, {first_name}"
    return full_name


# Test examples
print(format_name("John Smith"))      # Smith, John
print(format_name("Anita Rao"))       # Rao, Anita
```

```python
#task 4(4.2)
def count_vowels(string):
    """Count the number of vowels in a string."""
    vowels = "aeiouAEIOU"
    return sum(1 for char in string if char in vowels)


# Test cases
print(count_vowels("hello"))  # Output: 2
print(count_vowels("AI Assisted Coding"))  # Output: 7
```

```python
#task 5
def count_lines(filename):
    """
    Read a text file and count the total number of lines.
    
    Args:
        filename (str): The name of the file to read
        
    Returns:
        int: The total number of lines in the file
    """
    try:
        with open(filename, 'r') as file:
            line_count = sum(1 for line in file)
        return line_count
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return 0


# Example usage
if __name__ == "__main__":
    result = count_lines("file.txt")
    print(f"Total lines: {result}")
```
