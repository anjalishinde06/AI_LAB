```python

# Jupyter cell (index 0): Print Fibonacci sequence up to n terms without user-defined functions

try:
    n = int(input("Enter number of terms (n): "))
except Exception:
    print("Invalid input. Please enter an integer.")
else:
    if n <= 0:
        print("Please enter a positive integer.")
    else:
        a, b = 0, 1
        seq = []
        for _ in range(n):
            seq.append(str(a))
            a, b = b, a + b
        print("Fibonacci sequence:", " ".join(seq))
```
