**prompt:** Write a Python program to implement a function is_strong_password(password) that checks whether a password is strong based on given conditions (minimum 8 characters, includes uppercase, lowercase, digit, and special character, and no spaces), takes user input, prints the result, and includes assert test cases, unit tests, and doctests.

**Explanation:** This checks if a password is truly “strong” by making sure it has the right mix of characters and no spaces. The assert test cases, unit tests, and doctests ensure that the function works correctly in various scenarios.

```python
#Write a Python program to implement a function is_strong_password(password) that checks whether a password is strong based on given conditions (minimum 8 characters, includes uppercase, lowercase, digit, and special character, and no spaces), takes user input, prints the result, and includes assert test cases, unit tests, and doctests.
import re
import unittest
import doctest

def is_strong_password(password):
    """Check if the password is strong.

    >>> is_strong_password("StrongPass1!")
    True
    >>> is_strong_password("weakpass")
    False
    >>> is_strong_password("NoSpecialChar1")
    False
    """
    if not isinstance(password, str):
        return False
    if len(password) < 8:
        return False
    if ' ' in password:
        return False
    if not re.search(r'[A-Z]', password):
        return False
    if not re.search(r'[a-z]', password):
        return False
    if not re.search(r'[0-9]', password):
        return False
    if not re.search(r'[@$!%*?&]', password):
        return False
    return True


# ===== USER INPUT =====
password = input("Enter password: ")
if is_strong_password(password):
    print("Strong Password")
else:
    print("Weak Password")


# ===== ASSERT TEST CASES =====
assert is_strong_password("StrongPass1!") == True
assert is_strong_password("weakpass") == False
assert is_strong_password("Short1!") == False

print("Assert tests passed!")


# ===== UNIT TEST =====
class TestIsStrongPassword(unittest.TestCase):
    def test_strong(self):
        self.assertTrue(is_strong_password("StrongPass1!"))

    def test_weak(self):
        self.assertFalse(is_strong_password("weakpass"))


if __name__ == "__main__":
    print("\nRunning Unit Tests...")
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

    print("\nRunning Doctests...")
    doctest.testmod()

    print("\nAll tests executed!")
```

```python
#Write a Python program to generate at least 3 assert test cases for a classify_number(n) function. Implement using loops.Requirements,Classify numbers as Positive, Negative, or Zero.,Handle invalid inputs like strings and None.Include boundary conditions (-1, 0, 1).Example Assert Test Cases:assert classify_number(10) == "Positive"assert classify_number(-5) == "Negative"assert classify_number(0) == "Zero", takes user input, prints the result, and includes assert test cases, unit tests, and doctests.
def classify_number(n):
    """Classify a number as Positive, Negative, or Zero.

    >>> classify_number(10)
    'Positive'
    >>> classify_number(-5)
    'Negative'
    >>> classify_number(0)
    'Zero'
    >>> classify_number("string")
    'Invalid input'
    >>> classify_number(None)
    'Invalid input'
    """
    if not isinstance(n, (int, float)):
        return "Invalid input"
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"

# ===== USER INPUT =====
try:
    user_input = float(input("Enter a number: "))
    result = classify_number(user_input)
except ValueError:
    result = "Invalid input"
print(result)

# ===== ASSERT TEST CASES =====
assert classify_number(10) == "Positive"
assert classify_number(-5) == "Negative"
assert classify_number(0) == "Zero"
assert classify_number("string") == "Invalid input"
assert classify_number(None) == "Invalid input"
print("Assert tests passed!")

# ===== UNIT TEST =====
class TestClassifyNumber(unittest.TestCase):
    def test_positive(self):
        self.assertEqual(classify_number(10), "Positive")

    def test_negative(self):
        self.assertEqual(classify_number(-5), "Negative")

    def test_zero(self):
        self.assertEqual(classify_number(0), "Zero")

    def test_invalid_string(self):
        self.assertEqual(classify_number("string"), "Invalid input")

    def test_invalid_none(self):
        self.assertEqual(classify_number(None), "Invalid input")
if __name__ == "__main__":
    print("\nRunning Unit Tests...")
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

    print("\nRunning Doctests...")
    doctest.testmod()

    print("\nAll tests executed!")
```

**Explanation:** This function will check if two strings are anagrams, ignoring case, spaces, and punctuation. It also handles edge cases like empty strings and identical words to ensure it works in all scenarios.

```python
#Write a Python program to to generate at least 3 assert test cases formis_anagram(str1, str2) and implement the function.Requirements:Ignore case, spaces, and punctuation.Handle edge cases (empty strings, identical words).Example Assert Test Cases:assert is_anagram("listen", "silent") == True assert is_anagram("hello", "world") == False assert is_anagram("Dormitory", "Dirty Room") == True, Expected Output: • Function correctly identifying anagrams and passing all AI-generated tests. takes user input, prints the result, and includes assert test cases, unit tests, and doctests.
import re
import unittest
import doctest
def is_anagram(str1, str2):
    """Check if two strings are anagrams.

    >>> is_anagram("listen", "silent")
    True
    >>> is_anagram("hello", "world")
    False
    >>> is_anagram("Dormitory", "Dirty Room")
    True
    >>> is_anagram("", "")
    True
    >>> is_anagram("abc", "abc")
    True
    """
    def clean_string(s):
        return re.sub(r'[\W_]+', '', s).lower()

    return sorted(clean_string(str1)) == sorted(clean_string(str2))
# ===== USER INPUT =====
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")
if is_anagram(str1, str2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")

# ===== ASSERT TEST CASES =====
assert is_anagram("listen", "silent") == True
assert is_anagram("hello", "world") == False
assert is_anagram("Dormitory", "Dirty Room") == True
assert is_anagram("", "") == True
assert is_anagram("abc", "abc") == True
print("Assert tests passed!")
# ===== UNIT TEST =====
class TestIsAnagram(unittest.TestCase):
    def test_anagrams(self):
        self.assertTrue(is_anagram("listen", "silent"))
        self.assertTrue(is_anagram("Dormitory", "Dirty Room"))

    def test_not_anagrams(self):
        self.assertFalse(is_anagram("hello", "world"))

    def test_edge_cases(self):
        self.assertTrue(is_anagram("", ""))
        self.assertTrue(is_anagram("abc", "abc"))
if __name__ == "__main__":
    print("\nRunning Unit Tests...")
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

    print("\nRunning Doctests...")
    doctest.testmod()

    print("\nAll tests executed!")
```

**Prompt:** Write a Python program to  take any required input and generate at least 3 assert-based tests for an Inventory class with stock management.Methods:add_item(name, quantity),remove_item(name, quantity),get_stock(name),Example Assert Test Cases:inv = Inventory(),inv.add_item("Pen", 10),assert inv.get_stock("Pen") == 10,inv.remove_item("Pen", 5)assert inv.get_stock("Pen") == 5,inv.add_item("Book", 3),assert inv.get_stock("Book") == 3,the expected output: Fully functional class passing all assertions., takes user input, prints the result, and includes assert test cases, unit tests, and doctests.

**Explanation:** This Inventory class will allow you to manage stock by adding and removing items, as well as checking the current stock. The assert tests ensure that all methods work correctly and handle edge cases like removing more items than are in stock.

```python
#Write a Python program to  take any required input and generate at least 3 assert-based tests for an Inventory class with stock management.• Methods:add_item(name, quantity),remove_item(name, quantity),get_stock(name),Example Assert Test Cases:inv = Inventory(),inv.add_item("Pen", 10),assert inv.get_stock("Pen") == 10,inv.remove_item("Pen", 5)assert inv.get_stock("Pen") == 5,inv.add_item("Book", 3),assert inv.get_stock("Book") == 3,the expected output: Fully functional class passing all assertions., takes user input, prints the result, and includes assert test cases, unit tests, and doctests.
class Inventory:
    """A class to manage inventory stock."""

    def __init__(self):
        self.stock = {}

    def add_item(self, name, quantity):
        """Add items to the inventory."""
        if name in self.stock:
            self.stock[name] += quantity
        else:
            self.stock[name] = quantity

    def remove_item(self, name, quantity):
        """Remove items from the inventory."""
        if name in self.stock and self.stock[name] >= quantity:
            self.stock[name] -= quantity
        else:
            print(f"Cannot remove {quantity} of {name}. Not enough stock.")

    def get_stock(self, name):
        """Get the current stock of an item."""
        return self.stock.get(name, 0)
# ===== USER INPUT =====
inv = Inventory()
while True:
    action = input("Enter action (add/remove/get/exit): ").strip().lower()
    if action == "exit":
        break
    name = input("Enter item name: ").strip()
    if action in ["add", "remove"]:
        quantity = int(input("Enter quantity: "))
        if action == "add":
            inv.add_item(name, quantity)
        else:
            inv.remove_item(name, quantity)
    elif action == "get":
        print(f"Stock of {name}: {inv.get_stock(name)}")
    else:
        print("Invalid action. Please try again.")
# ===== ASSERT TEST CASES =====
inv = Inventory()
inv.add_item("Pen", 10)
assert inv.get_stock("Pen") == 10
inv.remove_item("Pen", 5)
assert inv.get_stock("Pen") == 5
inv.add_item("Book", 3)
assert inv.get_stock("Book") == 3
print("Assert tests passed!")
# ===== UNIT TEST =====
class TestInventory(unittest.TestCase):
    def setUp(self):
        self.inv = Inventory()

    def test_add_item(self):
        self.inv.add_item("Pen", 10)
        self.assertEqual(self.inv.get_stock("Pen"), 10)

    def test_remove_item(self):
        self.inv.add_item("Pen", 10)
        self.inv.remove_item("Pen", 5)
        self.assertEqual(self.inv.get_stock("Pen"), 5)

    def test_get_stock(self):
        self.inv.add_item("Book", 3)
        self.assertEqual(self.inv.get_stock("Book"), 3)
if __name__ == "__main__":
    print("\nRunning Unit Tests...")
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

    print("\nRunning Doctests...")
    doctest.testmod()

    print("\nAll tests executed!")
```

**Explanation:** This function will validate if a date string is in the correct format and represents a real date. If valid, it converts the date to a standardized format. The assert tests ensure that both valid and invalid inputs are handled correctly.

```python
#Write a Python program to generate at least 3 assert test cases for validate_and_format_date(date_str) to check and convert dates. • Requirements: Validate "MM/DD/YYYY" format,Handle invalid dates,Convert valid dates to "YYYY-MM-DD",Example Assert Test Cases:assert validate_and_format_date("10/15/2023") == "2023-10-15",assert validate_and_format_date("02/30/2023") == "Invalid Date",assert validate_and_format_date("01/01/2024") == "2024-01-01",Expected Output:Function passes all AI-generated assertions and handles edge cases.  takes user input, prints the result, and includes assert test cases, unit tests, and doctests.
from datetime import datetime
import unittest
import doctest
def validate_and_format_date(date_str):
    """Validate and format a date string.

    >>> validate_and_format_date("10/15/2023")
    '2023-10-15'
    >>> validate_and_format_date("02/30/2023")
    'Invalid Date'
    >>> validate_and_format_date("01/01/2024")
    '2024-01-01'
    >>> validate_and_format_date("invalid")
    'Invalid Date'
    """
    try:
        date_obj = datetime.strptime(date_str, "%m/%d/%Y")
        return date_obj.strftime("%Y-%m-%d")
    except ValueError:
        return "Invalid Date"
# ===== USER INPUT =====
date_input = input("Enter a date (MM/DD/YYYY): ")
result = validate_and_format_date(date_input)
print(result)
# ===== ASSERT TEST CASES =====
assert validate_and_format_date("10/15/2023") == "2023-10-15"
assert validate_and_format_date("02/30/2023") == "Invalid Date"
assert validate_and_format_date("01/01/2024") == "2024-01-01"
assert validate_and_format_date("invalid") == "Invalid Date"
print("Assert tests passed!")
# ===== UNIT TEST =====
class TestValidateAndFormatDate(unittest.TestCase):
    def test_valid_date(self):
        self.assertEqual(validate_and_format_date("10/15/2023"), "2023-10-15")

    def test_invalid_date(self):
        self.assertEqual(validate_and_format_date("02/30/2023"), "Invalid Date")

    def test_another_valid_date(self):
        self.assertEqual(validate_and_format_date("01/01/2024"), "2024-01-01")

    def test_invalid_format(self):
        self.assertEqual(validate_and_format_date("invalid"), "Invalid Date")
if __name__ == "__main__":
    print("\nRunning Unit Tests...")
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

    print("\nRunning Doctests...")
    doctest.testmod()

    print("\nAll tests executed!")

```
